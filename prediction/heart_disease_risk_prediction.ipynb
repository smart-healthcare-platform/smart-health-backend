# Fature Scaling and Building Neural Network model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint

# Target and feature separation
y = df['target']
X = df.drop('target', axis=1)

# Define which columns to scale
scaling_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
passthrough_cols = [col for col in X.columns if col not in scaling_cols]

# Apply StandardScaler to scaling columns only
scaler = StandardScaler()
X_scaled_part = scaler.fit_transform(X[scaling_cols])

# Convert scaled part back to DataFrame for merging
X_scaled_df = pd.DataFrame(X_scaled_part, columns=scaling_cols, index=X.index)

# Combine scaled and passthrough features
X_final = pd.concat([X_scaled_df, X[passthrough_cols].reset_index(drop=True)], axis=1)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

# Neural Network model
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))  # Binary classification

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Model checkpointing
checkpoint = ModelCheckpoint('heat_disease_prediction.h5', monitor='val_accuracy', mode='max', save_best_only=True, verbose=0)

# Fit the model
history = model.fit(X_train, y_train,
                    epochs=50,
                    batch_size=32,
                    validation_split=0.2,
                    callbacks=[checkpoint],
                    verbose=1)
```

# Prediction

```python
# Evaluate
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")

# Classification report
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))
```

```python
# Save the StandardScaler for use in the prediction service
import joblib

# Save the fitted scaler
joblib.dump(scaler, 'standard_scaler.pkl')
print("StandardScaler saved successfully!")

# Verify the saved scaler
loaded_scaler = joblib.load('standard_scaler.pkl')
print(f"Scaler loaded successfully. Mean values: {loaded_scaler.mean_}")
print(f"Scaler scale values: {loaded_scaler.scale_}")
```
